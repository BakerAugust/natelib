```python
import torch
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
from deep_learning.neurons import MultiLayerPerceptron
```

```python
BATCH_SIZE = 128
RANDOM_SEED = 88

torch.manual_seed(RANDOM_SEED)
```

```python
train_data = datasets.MNIST(root='data', train=True, transform=transforms.ToTensor())
test_data = datasets.MNIST(root='data', train=False, transform=transforms.ToTensor())

train_dataloader = DataLoader(train_data, batch_size=BATCH_SIZE, shuffle=True)
test_dataloader = DataLoader(test_data, batch_size=BATCH_SIZE, shuffle=False)
```

```python
mlp = MultiLayerPerceptron(
    n_features=28*28, 
    n_classes=len(train_data.classes),
    n_hidden=50)

optimizer = torch.optim.Adam
mlp.fit(dataloader=train_dataloader, learning_rate=0.01, epochs=100, optimizer=optimizer, verbose=True)
```

```python
## Calculate test accuracy
correct = 0
for x, y in test_data:
    logits = mlp(x)
    _, y_hat = torch.max(logits, dim=1)
    correct += torch.sum(y_hat == y)

print(correct/y_hat.size(0))
```
